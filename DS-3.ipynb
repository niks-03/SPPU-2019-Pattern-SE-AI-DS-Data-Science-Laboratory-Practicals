```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('employee_data.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0', 'id'], inplace=True)
```

```python
def age_group(age):
    if 11<= age <=20:
        return '11-20'
    if 21<= age <=30:
        return '21-30'
    if 31<= age <=40:
        return '31-40'
    if 41<= age <=50:
        return '41-50'
    if 51<= age <=60:
        return '51-60'
    if 61<= age <=70:
        return '61-70'
    else:
        return 'unknown'
```

```python
df['age_group'] = df['age'].apply(age_group)
```

```python
df.head()
```

```python
df.groupby(df['age_group'])[['salary','active_lifestyle']].mean()
```

```python
df.groupby(df['age_group'])[['salary']].median()
```

```python
df.groupby(df['age_group'])[['salary']].sum()
```

```python
df.groupby(df['age_group'])[['salary']].std()
```

```python
df.groupby(df['age_group'])[['salary']].min()
```
